```python
import re
import math

def isTimeFormat(time):
    timestamp_regex = r'\d{2}:\d{2}'
    return bool(re.match(timestamp_regex, time))
            
def convertTimeToNum(wakeupTime, sleepTime):   
    wakeupTime = re.split(':',wakeupTime)
    sleepTime = re.split(':',sleepTime)    
    
    for i in range(0,2):
        wakeupTime[i] = int(wakeupTime[i])
        sleepTime[i] = int(sleepTime[i])
        i += 1    
    
    wakeupTimeVal = wakeupTime[0] + (wakeupTime[1]/60)
    sleepTimeVal = sleepTime[0] + (sleepTime[1]/60)
    
    return wakeupTimeVal,sleepTimeVal

def convertNumToTime(mealInt):
    fractional, hoursInTime = math.modf(mealInt)
    minsInTime = int(fractional * 60 )
    if(minsInTime < 10):
        minsInTime = '0' + str(minsInTime)
    time = str(int(hoursInTime)) + ':' + str(minsInTime)
    print(time)


def calculateMealIntervals(wakeupTime, sleepTime, numOfMeals):
    wakeupTimeVal,sleepTimeVal = convertTimeToNum(wakeupTime,sleepTime)
    mealInterval = (sleepTimeVal - wakeupTimeVal) / numOfMeals
    newMealTime = wakeupTimeVal + 0.25
    count = 0
    while(newMealTime < sleepTimeVal and count < numOfMeals ):
        convertNumToTime(newMealTime)
        newMealTime += mealInterval 
        count += 1
              
def generateSchedules(wakeupTime,sleepTime, NoOfMeals):
    if(isTimeFormat(wakeupTime) and isTimeFormat(sleepTime)):
        calculateMealIntervals(wakeupTime, sleepTime, NoOfMeals)
        
    else:
           print("Invalid time format")

generateSchedules('11:30','20:20',3)
```
